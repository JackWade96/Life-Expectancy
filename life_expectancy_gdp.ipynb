```python
from matplotlib import pyplot as plt 
import pandas as pd
import seaborn as sns
```

```python
dp = pd.read_csv('all_data.csv')
print(dp.head())
```

```python
countrys = dp.Country
years = dp.Year
leb = dp['Life expectancy at birth (years)']
gdp = dp.GDP
```

```python
sns.barplot(x = countrys, y = gdp, data=dp)
plt.xticks(rotation = 30)
plt.show()
plt.clf()
```

```python
sns.barplot(x = countrys, y = leb, data = dp)
plt.xticks(rotation = 30)
plt.show()
plt.clf()
```

```python
f, ax = plt.subplots(figsize=(10, 15)) 
ax = sns.barplot(x="Country", y= leb, hue="Year", data=dp)
plt.ylabel('Life expectancy at birth in years')
plt.show()
```

```python

f, ax = plt.subplots(figsize=(10, 15)) 
ax = sns.barplot(data= dp, x='Country', y='GDP', hue='Year')
plt.xticks(rotation=90)
plt.ylabel('GDP in Trillions of U.S. Dollars')
plt.show()
```

```python
sns.violinplot(x = countrys, y = leb)
plt.xticks(rotation = 30)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=dp.Year, y=leb, hue=dp.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.ylabel("Life expectancy at birth (years)")
plt.show()
```
